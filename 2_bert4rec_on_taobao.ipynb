```python
from bert4rec_on_taobao import TaoBaoDataset,BERT4Rec,train,get_device,calculate_metrics_batch
```

```python
data_dir = 'bert4rec_on_taobao/data'
train_dataset = TaoBaoDataset(data_dir, max_len=20, split='train')
val_dataset = TaoBaoDataset(data_dir, max_len=20, split='val')
test_dataset = TaoBaoDataset(data_dir, max_len=20, split='test')
```

```python
print(len(train_dataset))
print(len(val_dataset)) #because of the leave-one-out method
print(len(test_dataset))
print(train_dataset.num_items)
print(train_dataset.mask_token)
```

## Hyperparameter Search

```python
class Config:
    #data_dir = os.path.join('data')
    model_save_dir = 'saved_models'

    # model hyperparameter
    max_len = 20
    hidden_units = 64
    num_heads = 4
    num_layers = 2
    dropout = 0.1

    # training parameters
    batch_size = 64
    learning_rate = 0.001
    num_epochs = 300
    eval_step = 1000
    patience = 5
    k_values = [1, 5, 10]

config = Config() # default settings
```

```python
model = BERT4Rec(
        item_num=train_dataset.num_items,
        max_len=config.max_len,
        hidden_units=config.hidden_units,
        num_heads=config.num_heads,
        num_layers=config.num_layers,
        dropout=config.dropout
    )
```

```python
device=get_device()
print(f'You are on {device}')
train(model,train_dataset,val_dataset,device,config)
```

# Final Model Training

```python
class Config:
    #data_dir = os.path.join('data')
    model_save_dir = 'saved_models'

    # model hyperparameter
    max_len = 20
    hidden_units = 32
    num_heads = 4
    num_layers = 2
    dropout = 0.1

    # training parameters
    batch_size = 64
    learning_rate = 0.001
    num_epochs = 300
    eval_step = 1000
    patience = 5
    k_values = [1, 5, 10, 50]

final_configurations = Config()
```

```python
model = BERT4Rec(
        item_num=train_dataset.num_items,
        max_len=final_configurations.max_len,
        hidden_units=final_configurations.hidden_units,
        num_heads=final_configurations.num_heads,
        num_layers=final_configurations.num_layers,
        dropout=final_configurations.dropout
    )
device=get_device()
print(f'You are on {device}')
train(model,train_dataset,val_dataset,device,final_configurations)
```

## Evaulation on Unseen data (testset)

```python
import torch
model.load_state_dict(torch.load('best_model.pth'))

model.eval()
val_hrs, val_ndcgs = calculate_metrics_batch(model, val_dataset,device=device,k_values=[1, 5, 10, 50])
test_hrs, test_ndcgs = calculate_metrics_batch(model, test_dataset,device=device,k_values=[1, 5, 10, 50])

# print results
print("\nValidation Metrics:")
for k in val_hrs.keys():
    print(f"HR@{k}: {val_hrs[k]:.4f}, NDCG@{k}: {val_ndcgs[k]:.4f}")
print("\nTest Metrics:")
for k in test_hrs.keys():
    print(f"HR@{k}: {test_hrs[k]:.4f}, NDCG@{k}: {test_ndcgs[k]:.4f}")
```
